```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading file

path='C://Users//TheGhost//Desktop//Hakathon//use.csv'
data=pd.read_csv(path)

#info about  data

a=data[['dispatch_day','rider_license_status','Number_of_Ratings' , 'target']]
a.head()
#data.shape

#Rescaling data - Normalization
a=(a-a.mean())/a.std()
a

a.insert(0,'Ones',1)
a

#separation of inputs and outputs
cols=a.shape[1]
print(cols)
x=a.iloc[ : , 0:cols-1]
y=a.iloc[ : , cols-1 : cols]
print(x)
print('-------')
print(y)

#Converting inputs and output into matrices
x=np.matrix(x)
y=np.matrix(y)
theta=np.matrix(np.array([0,0,0,0]))    #editable

print(theta.shape)
print(x.shape)
print(y.shape)
#print(x)


#gradient descent functions

def gradientDescent(X, y, theta, alpha, iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.ravel().shape[1])
    cost = np.zeros(iters)
    
    for i in range(iters):
        error = (X * theta.T) - y
        
        for j in range(parameters):
            term = np.multiply(error, X[:,j])
            temp[0,j] = theta[0,j] - ((alpha / len(X)) * np.sum(term))
            
        theta = temp
        cost[i] = computeCost(X, y, theta)
        
    return theta, cost

#------------------------------------------------------------

def computeCost(X, y, theta):
    inner = np.power(((X * theta.T) - y), 2)
    return np.sum(inner) / (2 * len(X))

#applying gradient descent

alpha=0.1
iters=1000000

# g: final thetas  - cost: matrix of the value of the Error in each itrate  - Error: final value of error
g,cost= gradientDescent(x,y,theta,alpha,iters)

Error=computeCost(x,y,g)

print(g)
print('--------------------------')
print(cost)
print('--------------------------')
print(Error)

# itrate=1000 - alpha=0.001  - Error=0.49869591936609337
# itrate=1000 - alpha=0.1    - Error=0.4984664686922686
# itrate=1000 - alpha=0.2    - Error=0.4984664686922686


# get best fit line for dispatch_day vs. target

x = np.linspace(a.l.min(), a.dispatch_day.max(), 100)
f = g[0, 0] + (g[0, 1] * x)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, f, 'r', label='Prediction')
ax.scatter(a.dispatch_day, a.dispatch_day, label='Training Data')
ax.legend(loc=2)
ax.set_xlabel('dispatch_day')
ax.set_ylabel('target')
ax.set_title('target vs. dispatch_day')
#--------------------------------------------------------------------------
# get best fit line for rider_license_status vs. target

x = np.linspace(a.rider_license_status.min(), a.rider_license_status.max(), 100)
f = g[0, 0] + (g[0, 1] * x)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, f, 'r', label='Prediction')
ax.scatter(a.rider_license_status, a.rider_license_status, label='Training Data')
ax.legend(loc=2)
ax.set_xlabel('rider_license_status')
ax.set_ylabel('target')
ax.set_title('target vs. rider_license_status')
#--------------------------------------------------------------------------
# get best fit line for Number of Ratings vs. target

x = np.linspace(a.rider_license_status.min(), a.rider_license_status.max(), 100)
f = g[0, 0] + (g[0, 1] * x)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, f, 'r', label='Prediction')
ax.scatter(a.rider_license_status, a.rider_license_status, label='Training Data')
ax.legend(loc=2)
ax.set_xlabel('rider_license_status')
ax.set_ylabel('target')
ax.set_title('target vs. rider_license_status')
```
